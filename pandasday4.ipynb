```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('sample_pivot.csv')
df.head()
```

```python
df_pv = pd.pivot_table(df,index = 'Region', columns= 'Type', values='Sales')
df_pv
```

```python
df_pv = pd.pivot_table(df,index = 'Region', columns= 'Type', values='Sales', aggfunc='max')
df_pv
```

```python
df_pv = pd.pivot_table(df,index = 'Region', columns= 'Type', values=['Sales','Units'])
df_pv
```
