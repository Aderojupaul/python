```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.style.use('ggplot')
sns.displot(x=tips['total_bill'], kde=True)
plt.show()
```

```python
plt.style.use('ggplot')
sns.regplot(x=tips['total_bill'],y =tips['tip'])
```

```python
sns.jointplot(x=tips['total_bill'],y =tips['tip'])
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'day')
```

```python
sns.countplot(x=tips['day'])
```

```python
sns.countplot(x=tips['day'],hue=tips['sex'],palette='autumn_r')
```

```python
sns.barplot(x=tips['day'],y = tips['total_bill'])
```

```python
sns.barplot(x=tips['day'],y = tips['total_bill'],estimator=min,hue=tips['time'])
```

```python
sns.violinplot(x=tips['time'],y = tips['tip'],estimator=max,hue=tips['sex'])
```

```python
sns.swarmplot(x=tips['smoker'],y = tips['total_bill'])
```

```python
sns.stripplot(x=tips['smoker'],y = tips['total_bill'])
```

```python
sns.boxplot(y = tips['total_bill'])
```

```python
sns.boxplot(x=tips['day'],y = tips['total_bill'])
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
df_pv = pd.pivot_table(df,index='year',columns='month',values='passengers')
df_pv
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df_pv,annot = True,fmt='0.5g')
```

```python
plt.figure(figsize=(15,10))
sns.clustermap(df_pv,annot = True,fmt='0.5g')
```

```python
df2 = tips.groupby('day').apply(sum)
df2.reset_index(inplace = True)
df2
```

```python
plt.figure(figsize=(6,6),dpi = 100)
plt.pie(df2['total_bill'],labels=df2['day'],autopct='%1.2f%%',shadow = True,explode=[0.0,0.0,0.0,0.2],startangle=90,colors = ['blue','red','green','yellow'])
plt.show()
```
