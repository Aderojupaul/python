```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('GDP.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
plt.plot(df['tyear'],df['Agriculture'])
```

```python
df = df[df['Period']== 'Annual']
plt.style.use('ggplot')
plt.plot(df['tyear'],df['Agriculture'], ls = 'dotted', lw = '3.5',marker = '+', markevery = 4, label = 'GDP')
plt.xlabel('YEAR')
plt.ylabel('AGRICULTURE')
plt.title('TREND OF AGRICULTURE GDP IN MILLIONS')
plt.legend()
plt.savefig('TREND.png')
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(df['tyear'],df['Agriculture'], ls = 'dotted', lw = '3.5',marker = '+', markevery = 4, label = 'AGRIC', color='green')
plt.plot(df['tyear'],df['Education'], ls = 'dotted', lw = '3.5',marker = '+', markevery = 4, label = 'EDUC', color = 'blue')
plt.plot(df['tyear'],df['CrudePetroleumAndNaturalGas'], ls = 'dotted', lw = '3.5',marker = '+', markevery = 4, label = 'CRUDE OIL', color='yellow')
plt.plot(df['tyear'],df['InformationAndCommunication'], ls = 'dotted', lw = '3.5',marker = '+', markevery = 4, label = 'ICT', color='red')
plt.xlabel('YEAR')
plt.ylabel('GDP')
plt.title('TREND OF VARIOUS INDUSTRIES GDP IN MILLIONS')
plt.legend(loc=[1,0])
plt.savefig('TREND2.png')
plt.show()
```

```python
fig = plt.figure(figsize=(15,15))
axes = fig.add_axes([0.1,0.1,0.5,0.3])
axes.bar(df['tyear'],df['Agriculture'],color = 'blue')
axes.set_xlabel('YEAR',fontsize = 35,color = 'blue')
axes.set_ylabel('GDP',fontsize = 35,color = 'blue')
axes.set_title('COMPARISM OF AGRICULTURAL GDP OVER THE YEAR', fontsize = 18, color = 'blue')
for label in axes.get_xticklabels():
    label.set_color('blue')
for label in axes.get_yticklabels():
    label.set_color('blue')
plt.show()
```
