```python
data1 = {
    'city':['Accra','Abuja','Daka','Lome'],
    'temperature':[45,37,35,20],
    'windspeed':[4,7,6,4]
}
table1 = pd.DataFrame(data1)

data2 = {
    'city':['Lagos','Kampala','Nairobo','Abuja','Daka'],
    'temperature':[30,25,36,37,28],
    'windspeed':[4,8,5,7,8]
}
table2= pd.DataFrame(data2)
```
